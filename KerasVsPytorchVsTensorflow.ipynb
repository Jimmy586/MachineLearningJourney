```python
# Importer les bibliothèques nécessaires
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import MinMaxScaler
from mlxtend.plotting import plot_confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import accuracy_score, classification_report
```

* Split X and Y(the target) then partially normalazing

```python
X = pd.get_dummies(df.drop(['Churn','Customer ID'], axis=1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
len(y[y == 0])
```

```python
len(y[y == 1])
```

# Balancing the data

```python
rus = RandomUnderSampler(random_state=2)
X_undersampled, y_undersampled = rus.fit_resample(X, y)
```

```python
len(y_undersampled[y_undersampled == 0])
```

```python
len(y_undersampled[y_undersampled == 1])
```

# Split to training set and testing set

```python
X_train, X_test, y_train, y_test = train_test_split(X_undersampled, y_undersampled, test_size=2)
```

```python
y_train.head()
```

# Deeplearning with tensorflow

importing the dependencies

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

creating the model

```python
model = Sequential()
# Ajouter une couche Dense avec 32 neurones et une activation ReLU(Rectified Linear Unit)
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))

# Ajouter une autre couche Dense avec 64 neurones et une activation ReLU
model.add(Dense(units=64, activation='relu'))
# Ajouter la couche de sortie Dense avec 1 neurone et une activation sigmoïde
model.add(Dense(units=1, activation='sigmoid'))



# metrics to minimize error rate
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

# train the model

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]

# Calculer la précision du modèle
accuracy = accuracy_score(y_test, y_hat)

accuracy
```

# Try with pytorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

create the model

```python
class Net(nn.Module):
    def __init__(self):
        """
        Initialise un modèle de réseau de neurones artificiels.

        Ce modèle a trois couches fully connected (entièrement connectées).

        Couches :
        - Couche 1 : 32 neurones avec une activation ReLU
        - Couche 2 : 64 neurones avec une activation ReLU
        - Couche 3 : 1 neurone avec une activation sigmoïde (pour la sortie binaire)
        """
        super(Net, self).__init__()
        self.fc1 = nn.Linear(len(X_train.columns), 32)
        self.fc2 = nn.Linear(32, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, x):
        """
        Effectue une propagation avant (forward pass) à travers le modèle.

        Args:
            x (torch.Tensor): Les données d'entrée.

        Returns:
            torch.Tensor: Les prédictions du modèle.
        """
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.sigmoid(self.fc3(x))
        return x
```

```python
# Créer une instance du modèle
model = Net()

# Définir la fonction de perte et l'optimiseur
criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
optimizer = optim.SGD(model.parameters(), lr=0.01)  # Stochastic Gradient Descent avec un taux d'apprentissage de 0.01
```

```python
def train(model, X_train, y_train, criterion, optimizer, epochs=200, batch_size=32):
    """
    Entraîne le modèle.

    Args:
        model (torch.nn.Module): Le modèle à entraîner.
        X_train (pd.DataFrame): Les données d'entraînement.
        y_train (pd.Series): Les étiquettes d'entraînement.
        criterion (torch.nn.Module): La fonction de perte.
        optimizer (torch.optim.Optimizer): L'optimiseur.
        epochs (int): Le nombre d'époques d'entraînement.
        batch_size (int): La taille du lot (mini-batch).

    Returns:
        None
    """
    for epoch in range(epochs):
        total_loss = 0.0
        for i in range(0, len(X_train), batch_size):
            inputs = torch.tensor(X_train[i:i+batch_size].values, dtype=torch.float32)
            labels = torch.tensor(y_train[i:i+batch_size].values, dtype=torch.float32).view(-1, 1)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            # Ajout pour afficher la perte à chaque étape
            total_loss += loss.item()

        # Afficher la perte et l'époque
        print(f'Époque [{epoch+1}/{epochs}], Perte: {total_loss}')

    # Message lorsque l'entraînement est terminé
    print('Entraînement terminé.')

# Exemple d'utilisation de la fonction train
# train(mon_modele, X_train, y_train, ma_fonction_de_perte, mon_optimiseur, epochs=200, batch_size=32)
```

```python
train(model, X_train, y_train, criterion, optimizer, epochs=200, batch_size=32)
```

```python
def predict(model, X_test):
    """
    Effectue des prédictions à l'aide du modèle.

    Args:
        model (torch.nn.Module): Le modèle entraîné.
        X_test (pd.DataFrame): Les données de test.

    Returns:
        torch.Tensor: Les prédictions du modèle.
    """
    with torch.no_grad():
        inputs = torch.tensor(X_test.values, dtype=torch.float32)
        outputs = model(inputs)
        predictions = (outputs >= 0.5).float()
    return predictions

# Faire des prédictions sur les données de test
y_hat = predict(model, X_test)
```

```python
# Calculer la précision du modèle
accuracy = (y_hat == torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)).float().mean()
accuracy
```
